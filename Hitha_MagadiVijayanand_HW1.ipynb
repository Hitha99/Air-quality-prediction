- c) Linear Regression Implementation Implement a linear regression model from scratch to regress the target variable,
Carbon monoxide (CO). (Remember: You can not use any libraries for the linear regression model.)
Some of the important formulaes used are listed below

- Linear Regression 
- Y = beta0 + beta1 * x1 + beta2 * x2 ...+ beta11 * x11 -> general equation for multiple linear regression(more than one feature)
- Or Y = Beta * X + bias ( where Beta = beta1, beta2 ... beta11)
- using gradient descent to fit the model gradient= (1/m) * [ x * (y - y_hat)] for bias = (1/m) * [ sum(y - y_hat)] -> sum of errors
- step 1: initialize the learning_rate lr and number of iterations n_iters and set beta(weight matrix) and bias to zero
- step 2: normalize the data
- step 3: use gradient descent and update the weight matrix(beta)
- step 4: write predict function using the updated/learnt weight using eq y = (x * beta) + bias
- step 5: find root mean squared error

### 1. Linear Regression Prediction Formula

The prediction in linear regression can be expressed as:

$$
\hat{y} = \beta_0 + \beta_1 x_1 + \beta_2 x_2 + \ldots + \beta_n x_n
$$

Where:
- y_hat is the predicted value.
- beta_0 is the intercept (bias term).
- beta_i are the coefficients for each feature x_i.

### 2. Normalization Formula

The normalization of feature \(x\) can be expressed as:

$$
x_{\text{norm}} = \frac{x - \mu}{\sigma}
$$

Where:
- x is the original feature value.
- mu is the mean of the feature.
- sigma is the standard deviation of the feature.
- m is the number of samples.

### 3. Gradient Descent Update Rules

The update rules for weights and bias during training using gradient descent can be expressed as:

$$
\beta_j := \beta_j - \alpha \frac{1}{m} \sum_{i=1}^{m} (y_{\text{pred}} - y_{\text{true}}) x_{ij}
$$

$$
\text{bias} := \text{bias} - \alpha \frac{1}{m} \sum_{i=1}^{m} (y_{\text{pred}} - y_{\text{true}})
$$

Where:
- beta_j is the weight for feature j.
- alpha is the learning rate.
- y_pred is the predicted value.
- y_true is the true value.
- x_ij is the value of feature \(j\) for the \(i\)-th sample.
- m is the number of samples.


```python
class LinearRegressionGD:
    # Linear Regression 
    # Y = beta0 + beta1 * x1 + beta2 * x2 ...+ beta11 * x11 -> general equation for multiple linear regression(more than one feature)
    # Or Y = Beta * X + bias ( where Beta = beta1, beta2 ... beta11)
    # using gradient descent to fit the model gradient= (1/m) * [ x * (y - y_hat)] for bias = (1/m) * [ sum(y - y_hat)] -> sum of errors
    # step 1: initialize the learning_rate lr and number of iterations n_iters and set beta(weight matrix) and bias to zero
    # step 2: normalize the data
    # step 3: use gradient descent and update the weight matrix(beta)
    # step 4: write predict function using the updated/learnt weight using eq y = (x * beta) + bias
    # step 5: find root mean squared error
    
    def __init__(self, lr=0.01, n_iters=10000):
        self.lr = lr 
        self.n_iters = n_iters 
        self.beta = None
        self.bias = None
        self.x_mean = None
        self.x_std = None

    # normalisation : most of the features contain large outliers which may skew the data hence we are normalising the data here
    def normalize(self,x):
        self.x_mean = np.mean(x, axis=0)
        self.x_std = np.std(x, axis=0)     
        x_norm = (x - self.x_mean) / self.x_std
        return x_norm

    # we pass the training data to get the trained weight/beta matrix and the bias
    def train(self,x_train,y_train):
        
        m, n = x_train.shape # m -> number_of_records in each feature | n -> number of features(here 11)
                
        self.beta = np.zeros(n).reshape(n,1) # intially the weight matrix/beta will be zero | reshaping to support the matrix multiplication
        self.bias = 0 # bias is zero initially

        x_norm = self.normalize(x_train) # x_norm -> normalized train values

        #gradient_descent
        for i in range(self.n_iters):
            y_pred = np.dot(x_norm,self.beta) + self.bias # y_hat = x * beta + bias general eq for y
            error = y_pred - y_train # error = predicted_values - actual values 
    
            d_beta = (1/ m) * np.dot(x_norm.T, error) # gradient for the weights
            d_bias = (1/ m) * np.sum(error) # gradient for the bias
     
            self.beta -= self.lr * d_beta # updating weights based on gradient
            self.bias -= self.lr * d_bias # updating bias based on gradient
            
    # finally predict the values using eq y = (x * beta) + bias
    def predict(self,x):
        x_norm = self.normalize(x)
        y_pred = np.dot(x_norm,self.beta) + self.bias        
        return y_pred
    
```

```python
x = features_data[features_list[1:]].to_numpy() # get all features except the target variable
y = features_data[['PT08.S1(CO)']].to_numpy() # target variable y

# splitting the data as train and test
test_size = 0.2 # 20% test | 80% train
split_index = int(x.shape[0] * (1 - test_size))
x_train, x_test, y_train, y_test = x[:split_index], x[split_index:], y[:split_index], y[split_index:]


linear_model = LinearRegressionGD()
#train the model
linear_model.train(x_train,y_train)

# prediction on the split test data
y_test_pred = linear_model.predict(x_test)

# root mean squared error
rmse= np.sqrt(np.mean((y_test-y_test_pred)**2))

#r_squared 
r_squared=1-(np.sum((y_test-y_test_pred)**2)/np.sum((y_test-np.mean(y_test))**2))

print(f'RMSE:{rmse}')
print(f'R-squared: {r_squared}')
```

```python
# getting the k fold indices
def kfold_indices(x, k):
    fold_size = len(x) // k
    indices = np.arange(len(x))
    folds = []
    for i in range(k):
        test_indices = indices[i * fold_size: (i + 1) * fold_size]
        train_indices = np.concatenate([indices[:i * fold_size], indices[(i + 1) * fold_size:]])
        folds.append((train_indices, test_indices))
    return folds

# since we are dooing k fold k=5
fold_indices = kfold_indices(x, 5)
```

- d) Implement a logistic regression model from scratch to predict this binary label. (Remember: You can not use any libraries for the logistic regression model.)
- Some new formulas used here other than gradient descent and normalization are BCE(binary cross entropy/ loss function here and sigmoid to scale values between 0 to 1)


**Loss Function:**

$$
\text{loss}(y_{\text{true}}, y_{\text{pred}}) = -\frac{1}{N} \sum_{i=1}^{N} \left[ y_{\text{true},i} \log(y_{\text{pred},i} + \epsilon) + (1 - y_{\text{true},i}) \log(1 - y_{\text{pred},i} + \epsilon) \right]
$$

Where:

$$
\epsilon = 10^{-9}
$$

**Sigmoid Function:**

$$
\text{sigmoid}(z) = \frac{1}{1 + e^{-z}}
$$



```python
class LogisticRegression:
    def __init__(self, lr=0.01, n_iters=10000):
        self.lr = lr 
        self.n_iters = n_iters 
        self.beta = None
        self.bias = None
        self.x_mean = None
        self.x_std = None
        self.losses = []

    # normalisation : most of the features contain large outliers which may skew the data hence we are normalising the data here
    def normalize(self,x):
        self.x_mean = np.mean(x, axis=0)
        self.x_std = np.std(x, axis=0)   
        x_norm = (x - self.x_mean) / ( self.x_mean - self.x_std)
        return x_norm

    def loss(self, y_true, y_pred):
        epsilon = 1e-9
        y_pred = np.clip(y_pred, epsilon, 1 - epsilon)
        y1 = y_true * np.log(y_pred + epsilon)
        y2 = (1-y_true) * np.log(1 - y_pred + epsilon)
        return -np.mean(y1 + y2)
        
    def sigmoid_fn(self, z):
        return 1/(1 + np.exp(-z))

    def train(self,x_train,y_train):
        m, n = x_train.shape
        
        self.beta = np.zeros(n).reshape(n,1) # intially the weight matrix/beta will be zero | reshaping to support the matrix multiplication
        self.bias = 0 # bias is zero initially

        x_norm = self.normalize(x_train) # x_norm -> normalized train values

        #gradient_descent
        for i in range(self.n_iters):
            y_pred = np.dot(x_norm,self.beta) + self.bias # y_hat = x * beta + bias general eq for y

            self.losses.append(self.loss(y_train,y_pred)) # computing and storing loss for each iteration
            
            d_sig = self.sigmoid_fn(y_pred) #applying the sigmoid function(takes any input and returns the value in between 0 to 1 -> non linear activation function

            error = d_sig - y_train # error = predicted_values - actual values 
    
            d_beta = (1/ m) * np.dot(x_norm.T, error) # gradient for the weights
            d_bias = (1/ m) * np.sum(error) # gradient for the bias
     
            self.beta -= self.lr * d_beta # updating weights based on gradient
            self.bias -= self.lr * d_bias # updating bias based on gradient

    def predict(self,x):
        x_norm = self.normalize(x)
        y_pred = np.dot(x_norm,self.beta) + self.bias
        y_sig = self.sigmoid_fn(y_pred) #unlike the linear reg here the value will be between 0 to 1 from sigmoid activation
        y_class = np.where(y_sig > 0.5, 1, 0) 
        return y_class

    def predict_probability(slef,x):
        x_norm = self.normalize(x)
        y_pred = np.dot(x_norm,self.beta) + self.bias
        y_pred_sig = self.sigmoid_fn(y_pred) #unlike the linear reg here the value will be between 0 to 1 from sigmoid activation
        return y_pred_sig

        
```

```python
def confusion_matrix(y, y_pred):
    actual = np.array(y)
    predicted = np.array(y_pred)

    # True Positives (TP): actual=1, predicted=1
    TP = np.sum((actual == 1) & (predicted == 1))
    
    # True Negatives (TN): actual=0, predicted=0
    TN = np.sum((actual == 0) & (predicted == 0))
    
    # False Positives (FP): actual=0, predicted=1
    FP = np.sum((actual == 0) & (predicted == 1))
    
    # False Negatives (FN): actual=1, predicted=0
    FN = np.sum((actual == 1) & (predicted == 0))
    
    return TP, TN, FP, FN
    
def accuracy(y,y_pred):    
    TP, TN, FP, FN = confusion_matrix(y,y_pred)   
    #Calculating accuracy using confusion matrix
    accuracy = (TP + TN)/(TP + TN + FP + FN)    
    return accuracy
    
def precision(y,y_pred):
    TP, TN, FP, FN = confusion_matrix(y,y_pred)  
    precision = TP/(TP + FP)
    return precision
    
def recall(y,pred):
    TP, TN, FP, FN = confusion_matrix(y,y_pred)
    recall = TP/(TP + FN)
    return recall

def f1_score(precision, recall):
    f1 = 2 * (precision * recall)/(precision + recall)
    return f1

acc = accuracy(y_test,y_pred)
prec = precision(y_test,y_pred)
rec = recall(y_test,y_pred)
f1 = f1_score(prec,rec)

print(f'Accuracy:{float(acc)}')
print(f'Precision:{float(prec)}')
print(f'Recall:{float(rec)}')
print(f'F1:{float(f1)}')
```
